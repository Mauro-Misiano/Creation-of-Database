```python
import random
import pandas as pd
from faker import Faker
import json

# Inicializar Faker con configuración en español
fake = Faker()
fake = Faker('es_ES')

# Tabla de Dimensiones

# Dimensión Estudiantes
estudiantes = []
for i in range(100):  # Generar 100 estudiantes
    estudiante = {
        "id_estudiante": i + 1,
        "nombre": fake.first_name(),
        "apellido": fake.last_name(),
        "edad": random.randint(18, 30),
        "sexo": random.choice(["masculino", "femenino"]),
        "carrera": random.choice(["Ingeniería", "Medicina", "Derecho", "Economía", "Psicología"]),
        "ciudad": fake.city(),
        "pais": fake.country()
    }
    estudiantes.append(estudiante)

# Dimensión Profesores
profesores = []
for i in range(20):  # Generar 20 profesores
    profesor = {
        "id_profesor": i + 1,
        "nombre": fake.first_name(),
        "apellido": fake.last_name(),
        "edad": random.randint(30, 65),
        "sexo": random.choice(["masculino", "femenino"]),
        "departamento": random.choice(["Ingeniería", "Medicina", "Derecho", "Economía", "Psicología"]),
        "ciudad": fake.city(),
        "pais": fake.country()
    }
    profesores.append(profesor)

# Dimensión Cursos
cursos = []
for i in range(10):  # Generar 10 cursos
    curso = {
        "id_curso": i + 1,
        "nombre": fake.word().capitalize(),
        "departamento": random.choice(["Ingeniería", "Medicina", "Derecho", "Economía", "Psicología"]),
        "horas_semanales": random.randint(2, 8)
    }
    cursos.append(curso)

# Tabla de Hechos

# Hechos Calificaciones
calificaciones = []
for i in range(500):  # Generar 500 registros de calificaciones
    calificacion = {
        "id_calificacion": i + 1,
        "id_estudiante": random.choice(estudiantes)["id_estudiante"],  # Selección de un ID de estudiante existente
        "id_profesor": random.choice(profesores)["id_profesor"],      # Selección de un ID de profesor existente
        "id_curso": random.choice(cursos)["id_curso"],                # Selección de un ID de curso existente
        "calificacion": round(random.uniform(0, 10), 2),
        "fecha": fake.date_between_dates(date_start=pd.to_datetime("2020-01-01"), date_end=pd.to_datetime("2023-12-31")).strftime('%Y-%m-%d')
    }
    calificaciones.append(calificacion)

# Guardar las dimensiones y los hechos en un archivo JSON
universidad = {
    "dimensiones": {
        "estudiantes": estudiantes,
        "profesores": profesores,
        "cursos": cursos
    },
    "hechos": {
        "calificaciones": calificaciones
    }
}

# Guardar el archivo JSON
with open('universidad.json', 'w') as f:
    json.dump(universidad, f, indent=4)

# Mensaje de confirmación
print("Generación de datos completada y guardada en universidad.json")

```
